```python
for key in state_dict.keys():
    f = open(''.join((key,'.txt')), "w")
    if  key[-4:] == 'bias':
        f.write(','.join((state_dict[key].numpy().astype('double').astype('str'))))
    else:
        for i in range(len(state_dict[key].numpy().astype('double').astype('str'))):
            f.write(','.join((state_dict[key].numpy().astype('double').astype('str')[i]))+'\n')
    f.close()
```
